```python
import matplotlib.pyplot as plt
from os import walk
import json
import numpy as np 
```

```python
mypath_cache = '/home/thais/Dev/TVMBench/tmp_logs/autoscheduler/llvm/old-cache/'
filenames_cache = next(walk(mypath_cache), (None, None, []))[2]  # [] if no file
```

```python
hashs_origin = []
values_origin = {}
for name in filenames_cache:
    if(".json" in name):
        f = open(mypath_cache + name, "r")
        best = 100000
        for line in f:
            result = json.loads(line)
            task = result['i'][0][0]
            r = sum(result['r'][0])/len(result['r'][0])
            if(task not in values_origin):
                values_origin[task] = []
            if(r < best):
                best = r
            if(r < 1000):
                values_origin[task].append(r)
```

```python
for task in values_origin:
    print(task)
```

```python
for idx, task in enumerate(values_origin):
    if(len(values_origin[task])):
        print(task)
        plt.figure(figsize=(25, 7))
        plt.plot(values_origin[task])
        plt.show()
```
